```python
import numpy as np
import astropy.units as u
import astropy.constants as c
import pandas as pd
import matplotlib.pyplot as plt
import scipy as sp
import scipy.interpolate

### =========  Function to do log-log interpolation =====================
def log_interp1d(xx, yy, kind='linear'):
    logx = np.log10(xx)
    logy = np.log10(yy)
    lin_interp = sp.interpolate.interp1d(logx, logy, kind=kind)
    log_interp = lambda zz: np.power(10.0, lin_interp(np.log10(zz)))
    return log_interp
```

### Perform unit conversion:
* **Cloudless** (Bobcat Spectra): $F_\nu$ (erg / cm^2 / s / Hz) 
* **Cloudy** (Diamondback Spectra): $F_\lambda$ (J / m^2 / s / m) 
* End result: Convert both to $F_\lambda$ (J / m^2 / s / m) (unit of Cloudy)
* Relation: $F_\lambda = F_\nu \frac{c}{\lambda^2}$

```python
# cloudlessPath ='~/Documents/GitHub/polar_vortice/data/spectras/bobcatCloudless_sp_t1200g1000nc_m0.00_co1.0_std.spec'
# cloudyPath = '~/Documents/GitHub/polar_vortice/data/spectras/diamondbackCloudy_sp_t1200g1000nc_m0.0_co1.0_std.spec'

cloudlessPath = '/Users/nguyendat/Documents/GitHub/polar_vortice/data/spectras/t1200g1000nc_m0.0_co1.0_std.spec'
cloudyPath = '/Users/nguyendat/Documents/GitHub/polar_vortice/data/spectras/t1200g1000f4_m0.0_co1.0_std.spec'

cloudlessSpec = pd.read_csv(cloudlessPath, sep='\s+', names = ['wave', 'flux'], skiprows=2)
cloudySpec = pd.read_csv(cloudyPath, sep='\s+', names = ['wave', 'flux'], skiprows=3)
```

```python
### Read dictionary, choose 25um print test
# micron
lamCloudy = cloudySpec.wave.copy() * u.micron
lamCloudless = cloudlessSpec.wave.copy() * u.micron
id1 = np.where(lamCloudy.values < 25 * u.micron)[0][0]
id2 = np.where(lamCloudless.values < 25 * u.micron)[0][0]

fluxCloudy_si = cloudySpec.flux.copy() * u.J/(u.m**2 * u.s * u.m) # (J / m^2 / s / m)
fluxCloudless_cgs = cloudlessSpec.flux.copy() * u.erg/((u.cm)**2 *u.s * u.Hz) # (erg / cm^2 / s / Hz) 

# print(id1, id2, lamCloudy.iloc[id1], lamCloudless.iloc[id2])
print('At 25 micron:')
print('fluxCloudy_lam_si = ', fluxCloudy_si.iloc[id1])
print('fluxCloudless_nu_cgs = ', fluxCloudless_cgs.iloc[id2])
```

```python
print(' Step 1: time c/lambda^2 (will take < 1min)') 
fluxCloudless_si_temp = [fluxCloudless_cgs[i] * c.c/lamCloudless[i]**2 for i in range(len(fluxCloudless_cgs))]
fluxCloudless_si_temp[id2]
```

```python
print(' Step 2: convert Cloudless to u.J/(u.m**2 * u.s * u.m) (will take < 1min)')
fluxCloudless_si = [val.to(u.J/(u.m**2 * u.s * u.m)) for val in fluxCloudless_si_temp]
```

```python
### 25 um post-conversion print test
# use this for output
fluxCloudless_unitless = [val.value for val in fluxCloudless_si]
fluxCloudy_unitless = [val.value for val in fluxCloudy_si]

print('Post-conversion:')
print('fluxCloudless_lam_si = ', fluxCloudless_si[id2])
print('At 25 micron:')
print('Cloudless/Cloudy ratio (flux_lam, si unit) = ', fluxCloudless_si[id2].value / fluxCloudy_si[id1].value)
```

```python
### plot the corrected unit spectra
%matplotlib inline
plt.close()

### First panels
fig, (ax1, ax2, ax3) = plt.subplots(1,3, figsize=(16,4), dpi=120)
ax1.plot(lamCloudy, fluxCloudy_si, ls='', alpha=0.15, marker='.', ms=0.1)
ax1.plot(lamCloudless, fluxCloudless_unitless, ls='', alpha=0.15, marker='.', ms=0.1)
ax1.plot([],[], color='tab:orange', label='Cloudless')
ax1.plot([],[], color='tab:blue', label='Cloudy')
ax1.set_xlim(0, 15)
ax1.set_ylabel('Flux density ($J/s/m^2/m$)')
ax1.set_xlabel('Wavelength ($\mu m$)')
ax1.legend()

ax2.plot(lamCloudy, fluxCloudy_si, ls='', alpha=0.15, marker='.', ms=0.5)
ax2.plot(lamCloudless, fluxCloudless_unitless, ls='', alpha=0.1, marker='.', ms=0.5)
ax2.plot([],[], color='tab:orange', label='Cloudless')
ax2.plot([],[], color='tab:blue', label='Cloudy')
ax2.set_xlim(1, 2)
ax2.set_xlabel('Wavelength ($\mu m$)')
ax2.legend()

ax3.plot(lamCloudy, fluxCloudy_si, ls='', alpha=0.15, marker='.', ms=1)
ax3.plot(lamCloudless, fluxCloudless_unitless, ls='', alpha=0.1, marker='.', ms=1)
ax3.plot([],[], color='tab:orange', label='Cloudless')
ax3.plot([],[], color='tab:blue', label='Cloudy')
ax3.set_xlim(1, 1.2)
ax3.set_xlabel('Wavelength ($\mu m$)')
ax3.legend()
```

```python
### Test the color from unit-corrected spectra
# J, H, K 
idJ1 = np.where(lamCloudy.values < 1.2 * u.micron)[0][0]
J1 = fluxCloudy_unitless[idJ1]
idH1 = np.where(lamCloudy.values < 1.6 * u.micron)[0][0]
H1 = fluxCloudy_unitless[idH1]
idK1 = np.where(lamCloudy.values < 2.2 * u.micron)[0][0]
K1 = fluxCloudy_unitless[idK1]

idJ2 = np.where(lamCloudless.values < 1.2 * u.micron)[0][0]
J2 = fluxCloudless_unitless[idJ2]
idH2 = np.where(lamCloudless.values < 1.6 * u.micron)[0][0]
H2 = fluxCloudless_unitless[idH2]
idK2 = np.where(lamCloudless.values < 2.2 * u.micron)[0][0]
K2 = fluxCloudless_unitless[idK2]

print('======= Color of cloudy & cloudless: ================')
print('cloudy [J-K]:', (J1-K1)/(J1-K1))
print('cloudless [J-K]:', (J2-K2)/(J1-K1))

print('\n')
print('cloudy [J-H]:', (J1-H1)/(J1-H1))
print('cloudless [J-H]:', (J2-H2)/(J1-H1))

print('\n')
print('======= Color ratio of cloudy & cloudless: ==========')
print('cloudy [J-K] / cloudless [J-K]:', (J1-K1)/(J2-K2))
print('cloudy [J-H] / cloudless [J-H]:', (J1-H1)/(J2-H2))
```

```python
### Test the silicate index from unit-corrected spectra
# sili, ca, cb
ids1 = np.where(lamCloudy.values < 9 * u.micron)[0][0]
s1 = fluxCloudy_unitless[ids1]
idca1 = np.where(lamCloudy.values < 6 * u.micron)[0][0]
ca1 = fluxCloudy_unitless[idca1]
idcb1 = np.where(lamCloudy.values < 13 * u.micron)[0][0]
cb1 = fluxCloudy_unitless[idcb1]

ids2 = np.where(lamCloudless.values < 9 * u.micron)[0][0]
s2 = fluxCloudless_unitless[ids2]
idca2 = np.where(lamCloudless.values < 6 * u.micron)[0][0]
ca2 = fluxCloudless_unitless[idca2]
idcb2 = np.where(lamCloudless.values < 13 * u.micron)[0][0]
cb2 = fluxCloudless_unitless[idcb2]

logY1 = log_interp1d((lamCloudy[idca1].value, lamCloudy[idcb1].value), (ca1, cb1), kind='linear')
logY2 = log_interp1d((lamCloudless[idca2].value, lamCloudless[idcb2].value), (ca2, cb2), kind='linear')

s1_intp = logY1(lamCloudy[ids1].value)
s2_intp = logY2(lamCloudless[ids2].value)

print('Silicate index;')
print('cloudy [si/continuum_interp]: ', s1/(s1_intp))
print('cloudless [si/continuum_interp]:', s2/(s2_intp))
```
